```python
import pandas as pd
import numpy as np; np.random.seed(8)
from numpy.polynomial.polynomial import polyfit
from matplotlib import pyplot as plt
import scipy.stats as stats
import seaborn as sns; sns.set(color_codes=True)
```

```python
fifa = pd.read_csv('CompleteDataset.csv')
fifa.head()
```

#### Our fifa data has uneven counts of null values per column, this is widely caused by the fact that Goal keepers play one position and one position only, so the other played positions are irrelevant to them, cuasing Nan values on the table dataframe.

#### so to fix this we will have to remove all the prefered positions for all players as for now the data in these columns is irrelevant to our goal. This will help us instead of dropping all the goal keepers' rows we will only remove none useful columns for now. 

```python
pos = fifa[['CAM','CB','CDM','CF','CM','ID','LAM','LB','LCB','LCM','LDM','LF','LM','LS','LW','LWB','RAM','RB','RCB','RCM','RDM','RF','RM','RS','RW','RWB','ST']] 
pos.head()
```

#  clean data
## lets clean the dataframe by removing rows with Nan values

```python
'''lets use drop() to remove all the irrelevant columns'''

fifap = fifa.drop(columns=pos)
fifap.info()
```

### lets remove columns that are not useful and alovate it to 'cfifa' (c=clean.) for a clean dataframe

```python
cfifa=fifap.drop(columns="Unnamed: 0")
cfifa.head()
```

```python
cfifa['Value'].head()
```

```python
def singi(string):
    if string[-1] == 'K':
        return float(string[1:-1])*1000
    elif string[-1] == 'M':
        return float(string[1:-1])*1000000
```

```python
cfifa['Value']= [singi(value) for value in cfifa['Value']]
cfifa['Wage'] = [singi(value) for value in cfifa['Wage']]
```

```python
cfifa['Value'].head()
```

### Value and Potential

```python
#plot a scatter betwwen Potential and Value to see the reltionship.

y = cfifa['Value']
x = cfifa['Potential']

plt.title("The relationship between value and potential")
plt.xlabel('Potential')
plt.ylabel('Value  (in a Million Euro)')
plt.scatter(x,y)

b, m = polyfit(x, y, 1)
plt.plot(x, y, '.')
plt.plot(x, b + m * x, '-')

plt.show()
```

```python
# calculate Pearson's correlation coefficient 
x = cfifa['Potential']
y = cfifa['Value']
slope, intercept, r, p, std_err = stats.linregress(x,y)

#std of potentetnial 
stdP = np.std(cfifa['Potential'])
print("standard deviation of Potential is",stdP)

#std of Value 
stdV = np.std(cfifa['Value'])
print("standard deviation of Value is",stdV)
```

## wage and age
plot to see if  Age determines the  Wage of the players

```python
# calculate Pearson's correlation coefficient 
x = cfifa['Wage']
y = cfifa['Age']
slope, intercept, r, p, std_err = stats.linregress(x,y)
```

```python
y = cfifa['Wage']
x = cfifa['Age']

plt.title("The relationship between value and potential")
plt.xlabel('Age')
plt.ylabel('Wage  (in a thousand Euro)')
plt.bar(x,y)
```
